```python
import matplotlib.pyplot as plt
import numpy as np

# Define the parameters
price_a = 3
price_b = 4
budget = 90

# Generate the x and y values for the budget line
x = np.linspace(0, budget/price_a, 100)
y = (budget - price_a*x) / price_b

# Define the utility function
utility = lambda a, b: a**0.5 * b**0.5

# Find the indifference curve that passes through the point (15, 15)
a = 15
b = (225 / a)**0.5
indifference_curve = lambda x: (225 / x)**0.5

# Generate the x and y values for the indifference curve
x_indifference = np.linspace(0.1, 30, 100)
y_indifference = indifference_curve(x_indifference)

# Plot the budget line and the indifference curve
plt.plot(x[x <= 40], y[x <= 40][y >= 0], label='Budget Line')
plt.xlim(0, 40)
plt.ylim(0, None)
plt.xlabel('Quantity of Good A')
plt.ylabel('Quantity of Good B')
plt.title('Consumer Budget Line and Indifference Curve')
plt.legend()
plt.grid()

# Find the equilibrium point
m = -price_a / price_b
x_eq = (budget / price_a) / (1 + m)
y_eq = m * x_eq
plt.plot(x_eq, y_eq, 'ro', label='Equilibrium Point')

plt.show()
```
